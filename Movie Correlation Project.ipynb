```python
import pandas as pd
import numpy as np
import seaborn as sns 

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
movie['budget'].fillna(0, inplace=True)
movie['budget'].replace([np.inf, -np.inf], 0, inplace=True)
movie['budget'] = movie['budget'].astype('int64')
```

```python
movie['gross'].fillna(0, inplace=True)
movie['gross'].replace([np.inf, -np.inf], 0, inplace=True)
movie['gross'] = movie['gross'].astype('int64')
```

```python
movie.isnull().sum()
```

```python
movie['released'] = movie['released'].astype(str)
def extract_year(date_string):
    try:
        # Split the string by comma and extract the year
        year = date_string.split(',')[1].strip().split()[0]
        return year
    except:
        # Handle any errors, such as invalid date formats
        return None

# Apply the custom function to create a new column 'year'
movie['correct_year'] = movie['released'].apply(extract_year)
```

```python
movie.head(20)
```

```python
movie.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
movie.drop_duplicates()
```

```python
nan_count = movie['company'].isna().sum()
print(nan_count)
```

```python
movie.dropna(subset=['company'], inplace=True)
```

```python
movie.shape
```

```python
movie.dropna(subset=['country', 'star','writer','rating'], inplace=True)
```

```python
movie.isnull().sum()
```

```python
# Calculate the mean for each column
score_mean = movie['score'].mean()
runtime_mean = movie['runtime'].mean()
votes_mean = movie['votes'].mean()

# Fill missing values with the calculated means
movie['score'].fillna(score_mean, inplace=True)
movie['runtime'].fillna(runtime_mean, inplace=True)
movie['votes'].fillna(votes_mean, inplace=True)
```

```python
mask = movie['correct_year'].isna()
movie.loc[mask, 'correct_year'] = movie.loc[mask, 'year']
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=movie['budget'], y=movie['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
#Plot budget vs gross using seanborn

sns.regplot(x='budget', y='gross', data=movie, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```
